### 1. Mouvement parabolique

```python
fichier=open('Donnees/data.txt','r')
fichier.readline()
lignes=fichier.readlines()
 
t=np.zeros(len(lignes))
x=np.zeros(len(lignes))
y=np.zeros(len(lignes))
 
for i in range(len(lignes)):
    ligne=lignes[i]
    ligne=ligne.replace(',','.')
    ligne=ligne.replace('E','e')
    t[i]=float(ligne.split(';')[0])
    x[i]=float(ligne.split(';')[1])
    y[i]=float(ligne.split(';')[3])
 
def derivee(x, t) :
    dx = []
    dt = t[1] - t[0]
 
    dx.append((x[1] - x[0]) / dt)
    for i in range(1, len(x) - 1) :
        dx.append((x[i+1] - x[i-1]) / (2  * dt))
    dx.append((x[-1] - x[-2]) / dt)
 
    return dx
 
# Parabole
plt.plot(x,y, color = 'red', linestyle = '--')
plt.show()
 
# D√©riv√©es
vx = derivee(x, t)
vy = derivee(y, t)

plt.plot(t, vx, color = 'blue')
plt.plot(t, vy, color = 'green')

plt.show()

# D√©riv√©es successives :

for i in range(1):
    vx = derivee(vx, t)
    vy = derivee(vy, t)

    plt.plot(t, vx, color = 'blue')
    plt.plot(t, vy, color = 'green')
    
    plt.show()
```

```python
pdy = np.polyfit(t, vy, 1)
pdx = np.polyfit(t, vx, 0)

print(f"g = {round(-pdy[0], 2)}")
print(f"v0 = {round(sqrt((pdx[0] ** 2) + (pdy[1] ** 2)), 2)}")
print(f"Œ∏ = {round(degrees(atan(pdy[1] / pdx[0])), 2)}")
```
